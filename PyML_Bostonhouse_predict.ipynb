```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pickle
from sklearn import metrics
```

```python
data_set = pd.read_csv("BostonHousing.csv")
```

```python
data_set.shape
```

```python
data_set
```

```python
x = data_set.drop("price", axis=1)
y = data_set['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
print(x_train, x_test, y_train, y_test)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model_prediction = model.predict(x_train)
```

```python
model_prediction
```

```python
s1 = metrics.r2_score(model_prediction, y_train)
```

```python
print("R2 score = ", s1)
```

```python
s2 = metrics.mean_absolute_error(model_prediction, y_train)
print("Mean absolute error score = ", s2)
```
